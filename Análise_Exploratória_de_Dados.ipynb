## Tratamento de Outliers

```python
houses_rent = pd.read_csv('https://raw.githubusercontent.com/matheusvazdata/aed-ada-santander/refs/heads/main/dados/houses_to_rent.csv')
houses_rent.head()
```

```python
houses_rent.describe()
```

```python
houses_rent['bathroom'].value_counts()
```

```python
# IQR: Intervalo Interquartil
q1 = houses_rent['bathroom'].quantile(0.25)
q3 = houses_rent['bathroom'].quantile(0.75)

IQR = q3 - q1
IQR
```

```python
houses_rent_outliers = houses_rent[(houses_rent['bathroom'] < q1 - 1.5 * IQR) | (houses_rent['bathroom'] > q3 + 1.5 * IQR)]
houses_rent_outliers
```

```python
houses_rent_inliers = houses_rent[~((houses_rent['bathroom'] < q1 - 1.5 * IQR) | (houses_rent['bathroom'] > q3 + 1.5 * IQR))]
houses_rent_inliers
```

```python
houses_rent.shape[0]
```

```python
houses_rent_outliers.shape[0] + houses_rent_inliers.shape[0]
```

```python
houses_rent_inliers.describe()
```

3) Vamos verificar agora se há outliers agora se há outliers

```python
# Visualizando distribuições com histogramas para ter intuição quanto a outliers
for col in df:

    sns.histplot(data=df, x=col, kde=True).set_title(f'Distribuição da variável "{col}"')
    plt.show()
```

```python
## Método do Quartil (IQR)

for col in df.drop(columns='quality'):

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    aux_outliers = df[(df[col] < Q1-(IQR*1.5))
                   | (df[col] > Q3+(IQR*1.5))]

    indices_outliers = aux_outliers.index.tolist()

    if len(indices_outliers) >= 1:

        print(f"A coluna {col} tem {len(indices_outliers)} outliers!")
        print("\nOs índices deles são:\n")
        print(indices_outliers)

    else:

        print(f"Acoluna {col} não tem outliers!")

    print()
    print("="*120)
    print()
```
