## Tratamento de Outliers

```python
df_3 = pd.read_csv('https://raw.githubusercontent.com/matheusvazdata/aed-ada-santander/refs/heads/main/dados/houses_to_rent.csv')
df_3.head()
```

```python
df_3.describe()
```

```python
df_3['bathroom'].value_counts()
```

```python
# IQR: Intervalo Interquartil
q1 = df_3['bathroom'].quantile(0.25)
q3 = df_3['bathroom'].quantile(0.75)

IQR = q3 - q1
IQR
```

```python
df_3_outliers = df_3[(df_3['bathroom'] < q1 - 1.5 * IQR) | (df_3['bathroom'] > q3 + 1.5 * IQR)]
df_3_outliers
```

```python
df_3_inliers = df_3[~((df_3['bathroom'] < q1 - 1.5 * IQR) | (df_3['bathroom'] > q3 + 1.5 * IQR))]
df_3_inliers
```

```python
df_3.shape[0]
```

```python
df_3_outliers.shape[0] + df_3_inliers.shape[0]
```

```python
df_3_inliers.describe()
```

```python
iris.head()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável "{col}"')
```

## Utilizando visualizações para séries históricas

```python
candy = pd.read_csv('https://raw.githubusercontent.com/matheusvazdata/aed-ada-santander/refs/heads/main/dados/candy_production.csv')
candy.head(12)
```

```python
candy.tail(12)
```

```python
candy.info()
```

```python
candy['observation_date'] = pd.to_datetime(candy['observation_date'])
candy.info()
```

```python
candy.plot(x='observation_date', y='industrial_production')
```

```python
candy_filtered = candy[candy['observation_date'] >= '2010-01-01']
candy_filtered
```

```python
ax = candy_filtered.plot(x='observation_date', y='industrial_production', figsize=(14,5))
```

```python
ax = candy_filtered.plot(x='observation_date', y='industrial_production', figsize=(14,5))
xcoords = [f'{year}-01-01' for year in range(2011, 2018)]

for xc in xcoords:
  plt.axvline(x=xc, color='silver', linestyle='--')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

candy_filtered.set_index('observation_date', inplace=True)

analysis = candy_filtered[['industrial_production']].copy()

decompose_result = seasonal_decompose(analysis, model='multiplicative')

trend = decompose_result.trend
easonal = decompose_result.seasonal
residual = decompose_result.resid

decompose_result.plot()
```
