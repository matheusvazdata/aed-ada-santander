### Pergunta de negócio: qual a culinária mais bem avaliada na região do Central Park, NY?

```python
df.Location.value_counts()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns.to_list()
numeric_cols
```

```python
df[numeric_cols]
```

```python
df.groupby(['Location'])[numeric_cols].mean()
```

```python
df.groupby(['Location', 'Cuisines'])[numeric_cols].mean()
```

```python
df.Location = df.Location.replace('Central Park,ny', 'Central Park,NY')
df.groupby(['Location', 'Cuisines'])[numeric_cols].mean()
```

```python
df.Location.value_counts()
```

```python
df[df['Location'] == 'Central Park,NY'].Cuisines.value_counts()
```
